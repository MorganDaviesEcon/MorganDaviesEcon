```python
import pandas as pd 
```

```python
# read csv file 
df = pd.read_csv('electoral_data.csv',skiprows=3)
```

```python
# checking the data frame
df.head(80)
```

```python
# only want specific columns
df2 = df[['Unnamed: 1','Unnamed: 5']]
```

```python
df2.head(40)
```

```python
# renaming the columns
df3 = df2.rename(columns={'Unnamed: 1':'constituency','Unnamed: 5':'electorate'})
```

```python
df3
```

```python
# change the index
df4 = df3.set_index('constituency')
```

```python
# want the data for the regions of ireland, not the other random numbers
df5 = df4.loc[{'BELFAST EAST','BELFAST WEST','BELFAST NORTH','BELFAST SOUTH','EAST ANTRIM','EAST LONDONDERRY','FERMANAGH AND SOUTH TYRONE','FOYLE','LAGAN VALLEY','MID ULSTER','NEWRY AND ARMAGH','NORTH ANTRIM','NORTH DOWN','SOUTH ANTRIM','SOUTH DOWN','STRANGFORD','UPPER BANN','WEST TYRONE'}]
```

```python
df5.to_csv('northern_ireland_electoral_new.csv')
```
