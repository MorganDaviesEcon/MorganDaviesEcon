```python
# import usuals
import requests
import pandas as pd
import os
```

```python
# put data files into colab workbook
Unemploy = pd.read_csv('https://raw.githubusercontent.com/MorganDaviesEcon/MorganDaviesEcon.github.io/main/US_unemploy.csv')
Job = pd.read_csv('https://raw.githubusercontent.com/MorganDaviesEcon/MorganDaviesEcon.github.io/main/US_job_open.csv')
```

```python
# checking data frame
Unemploy
```

```python
data = pd.DataFrame(Unemploy)
```

```python
data
```

```python
# making the dates into a more readable format
Unemploy_new = data.replace(to_replace="/",value="-", regex=True)
```

```python
Unemploy_new
```

```python
# making the dates into a more readable format
Job_new = Job.replace(to_replace="/",value="-", regex=True)
Job_new
```

```python
df = pd.DataFrame(Unemploy_new)
df
```

```python
# merge the two data frames together
df = df.merge(Job_new)
```

```python
df
```

```python
# rename columns
df2 = df.rename(columns={'UNEMPLOY':'Unemployment_Level',"JTSJOL":"Job_Openings"})
```
