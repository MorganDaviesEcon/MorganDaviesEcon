```python
dfs=[]
for x in ['LVS','WYNN','MGM','MLCO','PENN','PDYPY','GVC.VI','PDYPF']:
  df = yf.Ticker(x).history(period='1y')
  df=df[['Close']]
  df['Name']=x
  dfs.append(df)
```

```python
dfs[0].head(10)
```

```python
casino_stocks = pd.concat(dfs)
```

```python
casino_stocks_new = casino_stocks.dropna()
```
