```python
import requests, os, pandas as pd
```

```python
df = pd.read_csv('API_HD.HCI.OVRL_DS63_en_csv_v2_4523493.csv',skiprows=4)
df
```

```python
d2 = df[['Country Name','Country Code','2020']]
```

```python
d2
```

```python
df3 = d2.dropna().reset_index(drop=True)
```

```python
df3.rename(columns={'2020':'HDI'},inplace=True)
```

```python
df3
```

```python
gdp = pd.read_csv('API_NY.GDP.PCAP.CD_DS2_en_csv_v2_4701206.csv',skiprows=4)
gdp
```

```python
gdp1 = gdp[['Country Code','2020']]
```

```python
gdp1
```

```python
merge = pd.merge(df3,gdp1,on='Country Code')
```

```python
merge
```

```python
merge_clean = merge.dropna().reset_index(drop=True)
```

```python
merge_clean
```

```python
merge_clean.rename(columns={'2020':'GDPpc'},inplace=True)
```

```python
merge_clean2 = merge_clean[['HDI','GDPpc']].round(decimals=2)
```

```python
merge_clean2.rename(columns={'HDI':'HDIreal','GDPpc':'GDPpcreal'},inplace=True)
```

```python
merge_clean2
```

```python
merge_clean['HDI'] = merge_clean2['HDI']
```

```python
merge_clean['GDPpc'] = merge_clean2['GDPpc']
```

```python
merge_clean
```

```python
merge_clean.to_csv('GDP_HDI.csv')
```
