```python
import pandas as pd
```

```python
df = pd.read_csv('Clean_Water.csv',skiprows=3)
```

```python
df.head(2)
```

```python
# Drop columns we don't need
df2 =df.drop(['Indicator Name','Indicator Code'],axis=1)
```

```python
df2
```

```python
# stack the remaining columns to make it TIDY format
df3 = df2.set_index(['Country Name','Country Code']).stack().reset_index()
```

```python
df3
```

```python
# check columns
df3.columns
```

```python
# rename columns
df4 = df3.rename(columns={'level_2':'Year',0:'Using Safe Water (Percent of Population)'})
```

```python
df4
```

```python
# changing index to country code
df5 = df4.set_index('Country Code')
```

```python
df5
```

```python
# get specific country data
df6 = df5.loc[{'ZWE','WLD','TUN','SVN','BLR','ECU'}]
```

```python
df6
```

```python
df7 = df6.reset_index()
```

```python
# check we have all the data needed from the specific countries
exported_values = list( df7.T.to_dict().values())
```

```python
exported_values
```

```python
import json
```

```python
# export into JSON form to upload to vega
open('clean_water.json','w').write(json.dumps(exported_values))
```
